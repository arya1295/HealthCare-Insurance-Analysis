```python
'''The gender of the patient may be an important factor in determining the cost of hospitalization.
The salutations in a beneficiary's name can be used to determine their gender. Make a new field for the beneficiary's gender.'''
```

```python
def get_gender(name):
    # List of known salutations
    salutations = ['Mr.', 'Ms.', 'Mrs.', 'Miss', 'Dr.']
    # Split the name into parts
    name_parts = name.split()
    # Initialize gender as 'Unknown'
    gender = 'Unknown'
    # Iterate over name parts to find a salutation
    for part in name_parts:
        if part in salutations:
            # Map the salutation to gender
            if part == 'Mr.':
                gender = 'Male'
            elif part in ['Ms.', 'Mrs.', 'Miss']:
                gender = 'Female'
            elif part == 'Dr.':
                gender = 'Unknown'  # 'Dr.' could be either, so we use 'Unknown'
            break  # Stop after finding the first salutation
    return gender

# Apply the function to the 'name' column and create a new 'gender' column
df_final1['gender'] = df_final1['name'].apply(get_gender)
```

Test the following null hypotheses:

a.The average hospitalization costs for the three types of hospitals are not significantly different.

Null Hypothesis(H0) : The average hospitalization costs for the three types of hospitals are not significantly different.

Alternate Hypothesis(H1) : The average hospitalization costs for at least one hospital tier are significantly different from the others.

```python
hc1 = df_final[['charges','Hospital tier']]
```

```python
hc1.rename(columns={'Hospital tier':'Hospital_tier'},inplace= True)
```

```python
hc1.head()
```

```python
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
model = ols('charges ~ C(Hospital_tier)', data=hc1).fit()
```

```python
anova_table = sm.stats.anova_lm(model, typ=2)
```

```python
print(anova_table)
```

If the Significance Level is 0.05, 0.01, or even 0.001 we see that the P-value is **4.258597e-179** which is far less than any common significance level.


**Decision** : We should **reject** the null hypothesis (H₀) and accept the alternative hypothesis (H₁), which states that there are significant differences in the average hospitalization costs among the different hospital tiers.

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.preprocessing import StandardScaler

X = df_final2.drop('charges', axis=1)
y = df_final2['charges']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define k-fold cross validation
kf = KFold(n_splits=5, shuffle=True, random_state=1)

# Linear Regression
linear_model = LinearRegression()
linear_scores = cross_val_score(linear_model, X_scaled, y, scoring='neg_mean_squared_error', cv=kf)

# Ridge Regression with alpha=1.0
ridge_model = Ridge(alpha=1.0)
ridge_scores = cross_val_score(ridge_model, X_scaled, y, scoring='neg_mean_squared_error', cv=kf)

# Convert negative MSE to positive
linear_rmse_scores = np.sqrt(-linear_scores)
ridge_rmse_scores = np.sqrt(-ridge_scores)

print(f"Linear Regression RMSE: {linear_rmse_scores.mean()} (std: {linear_rmse_scores.std()})")
print(f"Ridge Regression RMSE: {ridge_rmse_scores.mean()} (std: {ridge_rmse_scores.std()})")
```

Implement the stratified 5-fold cross validation technique for both model building and validation

```python
from sklearn.model_selection import StratifiedKFold
X = df_final2.drop('charges', axis=1)
y = df_final2['charges']

# Bin the target variable
num_bins = 5
y_binned = pd.cut(y, bins=num_bins, labels=False)

X = X.reset_index(drop=True)
y = y.reset_index(drop=True)
y_binned = y_binned.reset_index(drop=True)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define stratified k-fold cross validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)

# Function to evaluate model with stratified k-fold cross-validation
def evaluate_model(model, X, y, y_binned):
    scores = []
    for train_index, test_index in skf.split(X, y_binned):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        model.fit(X_train, y_train)
        score = model.score(X_test, y_test)
        scores.append(score)
    return np.array(scores)

# Linear Regression
linear_model = LinearRegression()
linear_scores = evaluate_model(linear_model, X_scaled, y, y_binned)

# Ridge Regression
ridge_model = Ridge(alpha=1.0)
ridge_scores = evaluate_model(ridge_model, X_scaled, y, y_binned)

print(f"Linear Regression R^2: {linear_scores.mean()} (std: {linear_scores.std()})")
print(f"Ridge Regression R^2: {ridge_scores.mean()} (std: {ridge_scores.std()})")
```

Utilize effective standardization techniques and hyperparameter tuning

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.metrics import make_scorer, mean_squared_error

X = df_final2.drop('charges', axis=1)
y = df_final2['charges']

# Bin the target variable (for stratification purposes)
num_bins = 5
y_binned = pd.cut(y, bins=num_bins, labels=False)

# Define a custom scorer for GridSearchCV
def rmsle(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

rmsle_scorer = make_scorer(rmsle, greater_is_better=False)

# Define standard scaler and ridge parameters
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', Ridge())
])

param_grid = {
    'regressor__alpha': [0.1, 0.01,1.0,0.5,0.05,10.0]
}

# GridSearchCV for Ridge Regression
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring=rmsle_scorer)
grid_search.fit(X, y)

print(f"Best Parameters for Ridge Regression: {grid_search.best_params_}")
print(f"Best RMSLE Score: {-grid_search.best_score_}")

# Linear Regression (no hyperparameters to tune, but we can still use GridSearchCV for consistency)
linear_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression())
])

linear_scores = cross_val_score(linear_pipeline, X, y, cv=5, scoring=rmsle_scorer)
print(f"Linear Regression RMSLE: {-linear_scores.mean()} (std: {linear_scores.std()})")
```

```python
from sklearn.linear_model import Ridge, Lasso
X = df_final2.drop('charges', axis=1)
y = df_final2['charges']

def rmsle(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

rmsle_scorer = make_scorer(rmsle, greater_is_better=False)

# Define pipelines for Ridge and Lasso Regression
pipeline_ridge = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', Ridge())
])

pipeline_lasso = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', Lasso())
])

# Define parameter grids for Ridge and Lasso
param_grid_ridge = {
    'regressor__alpha': [0.1, 1.0, 10.0, 100.0]
}

param_grid_lasso = {
    'regressor__alpha': [0.1, 1.0, 10.0, 100.0]
}

# Setup KFold for cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=1)

# GridSearchCV for Ridge Regression
grid_search_ridge = GridSearchCV(pipeline_ridge, param_grid_ridge, cv=kf, scoring=rmsle_scorer)
grid_search_ridge.fit(X, y)

print(f"Best Parameters for Ridge Regression: {grid_search_ridge.best_params_}")
print(f"Best RMSLE Score for Ridge Regression: {-grid_search_ridge.best_score_}")

# GridSearchCV for Lasso Regression
grid_search_lasso = GridSearchCV(pipeline_lasso, param_grid_lasso, cv=kf, scoring=rmsle_scorer)
grid_search_lasso.fit(X, y)

print(f"Best Parameters for Lasso Regression: {grid_search_lasso.best_params_}")
print(f"Best RMSLE Score for Lasso Regression: {-grid_search_lasso.best_score_}")
```

Develop Gradient Boost model and determine the variable importance scores,and identify the redundant variables

```python
from sklearn.ensemble import GradientBoostingRegressor
X = df_final2.drop('charges', axis=1)
y = df_final2['charges']

# Initialize Gradient Boosting Regressor
gb_regressor = GradientBoostingRegressor()

# Set up pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Standardize features
    ('model', gb_regressor)         # Gradient Boosting Regressor
])

# Initialize KFold for cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=1)

# Fit the model using cross-validation
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Fit the model
    pipeline.fit(X_train, y_train)

    # Predict and evaluate
    print(f"Fold results: {pipeline.score(X_test, y_test)}")

# Fit the model on the entire dataset
pipeline.fit(X, y)

# Get feature importances
importances = pipeline.named_steps['model'].feature_importances_

# Create a DataFrame for feature importances
feature_importances = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print("Feature Importances:")
print(feature_importances)

# Identify redundant variables (those with low importance scores)
# Define a threshold to filter out redundant variables
threshold = 0.01
redundant_features = feature_importances[feature_importances['Importance'] < threshold]

print("Redundant Features:")
print(redundant_features)
```
